```python
# Create code to answer each of the following questions.
# Hint: You will need multiple target URLs and multiple API requests.

# Dependencies
import time
import requests
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
data_name = 'historical_covid_data'
csv = '.csv'
url = "https://covidtracking.com/api/v1/states/daily.json"
timestr = time.strftime("%Y%m%d")
```

```python
historical_covid_data_latest_json = requests.get(url).json()
```

```python
new_df = pd.DataFrame(historical_covid_data_latest_json)
```

```python
date_list = new_df['date'].unique()
date_list[0]
```

```python
for date in range(0, len(date_list)-4):
    print(date_list[date])
```

```python
#setup new columns
#The column I have that calculates hospital cases is based on the cumulative hospitalizations.
#some of those columns are 0.  We will then use the difference of the current hospitalizations to fill the value.
#new hospitalized cases based on difference from current cases
new_df['hosp_increase_hospitalizedCurrent'] = 0
#new hospitalized cases based on whether we use the existing column 'hospitalizedIncrease' or the new column
#which is 'hosp_increase_hospitalized_final'
new_df['hosp_weekly_increase_hospitalized_final'] = 0
#marked True if hospitalizations must be calculated using the current hospitalization
new_df['use_current_hospitalizations'] = False
new_df['new_pos_weekly'] = 0
new_df['new_hos_weekly'] = 0
new_df['new_hos_weekly_current'] = 0
new_df['trend_cases'] = 0
new_df['trend_rate_cases'] = 0
new_df['trend_hosp'] = 0
new_df['trend_rate_hosp'] = 0
```

```python
state_list = new_df['state'].unique()
state_list
```

```python
#determine states that have no values in the hospitalizedCumulative column
for state in range(0,len(state_list)-1):
    state_sum = new_df.loc[(new_df['state'] == state_list[state] ), 'hospitalizedCumulative'].sum()
    print(state_list[state], '  ', state_sum)
    if state_sum == 0:
        new_df.loc[(new_df['state'] == state_list[state] ), 'use_current_hospitalizations'] = True
```

```python
# need to populate the two new hosp_increase columns in cell above.
for date in range(0,30):
    for state in range(0,len(state_list)-1):
        hosp_current = new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] == date_list[date]) ), 'hospitalizedCurrently'].values[0]
        hosp_current_day_less_1 = new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] == date_list[date+1]) ), 'hospitalizedCurrently'].values[0]
        hosp_current_diff = hosp_current - hosp_current_day_less_1
        print(state_list[state],'  ', date_list[date], '  ', date_list[date+1], '  ', hosp_current, '  ', hosp_current_day_less_1, '  ', hosp_current_diff)
        new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] == date_list[date]) ), 'hosp_increase_hospitalizedCurrent'] = hosp_current_diff
```

```python
#instead of doing all past dates, just do past month  -   len(date_list)-4
for date in range(0, 30):
    for state in range(0, len(state_list)-1):
        sum_week_cases = new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] <= date_list[date]) ), 'positiveIncrease'][0:7].sum()
        sum_week_hospitalized = new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] <= date_list[date]) ), 'hospitalizedIncrease'][0:7].sum()
        sum_week_hospitalized_current = new_df.loc[((new_df['state'] == state_list[state])& (new_df['date'] <= date_list[date]) ), 'hosp_increase_hospitalizedCurrent'][0:7].sum()

        new_df.loc[((new_df['state'] == state_list[state]) & (new_df['date'] == date_list[date])), 'new_pos_weekly'] = sum_week_cases
        new_df.loc[((new_df['state'] == state_list[state]) & (new_df['date'] == date_list[date])), 'new_hos_weekly'] = sum_week_hospitalized
        new_df.loc[((new_df['state'] == state_list[state]) & (new_df['date'] == date_list[date])), 'new_hos_weekly_current'] = sum_week_hospitalized_current
        if new_df.loc[new_df['state'] == state_list[state], 'use_current_hospitalizations'].any():
            new_df.loc[((new_df['state'] == state_list[state]) & (new_df['date'] == date_list[date])),'hosp_weekly_increase_hospitalized_final'] = sum_week_hospitalized_current
        else:
            new_df.loc[((new_df['state'] == state_list[state]) & (new_df['date'] == date_list[date])),'hosp_weekly_increase_hospitalized_final'] = sum_week_hospitalized
```

```python
for state in range(0, len(state_list)-1):
        current_total_week_cases = new_df.loc[((new_df['state'] == state_list[state])), 'new_pos_weekly'].values[0]
        previous_total_week_cases = new_df.loc[((new_df['state'] == state_list[state]) ), 'new_pos_weekly'].values[7]
        month_avg_week_cases = new_df.loc[((new_df['state'] == state_list[state]) ), 'new_pos_weekly'][7:35].mean()
        if not(previous_total_week_cases == 0):
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_rate_cases'] = current_total_week_cases / previous_total_week_cases
        if current_total_week_cases > previous_total_week_cases:
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_cases'] = "up"
        else:
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_cases'] = "down"
```

```python
for state in range(0, len(state_list)-1):
        current_total_week_hosp = new_df.loc[((new_df['state'] == state_list[state])), 'hosp_weekly_increase_hospitalized_final'].values[0]
        previous_total_week_hosp = new_df.loc[((new_df['state'] == state_list[state]) ), 'hosp_weekly_increase_hospitalized_final'].values[7]
        month_avg_week_hosp = new_df.loc[((new_df['state'] == state_list[state]) ), 'hosp_weekly_increase_hospitalized_final'][7:35].mean()
        if not(previous_total_week_hosp == 0):
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_rate_hosp'] = current_total_week_hosp / previous_total_week_hosp
        if current_total_week_cases > previous_total_week_cases:
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_hosp'] = "up"
        else:
            new_df.loc[((new_df['state'] == state_list[state]) ), 'trend_hosp'] = "down"
```
